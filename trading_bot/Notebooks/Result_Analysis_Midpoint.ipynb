### Load Results

```python
midpoint_disc_seed1 = pd.read_csv('../Results/resultsimple_buy_sell_at_midpointaccb_nosec_bid_askdiscrete2020-08-09 12:56:17.190484.csv')
midpoint_disc_seed2 = pd.read_csv('../Results/resultsimple_buy_sell_at_midpointaccb_nosec_bid_askdiscrete2020-08-09 12:21:06.657610.csv')
midpoint_cont_seed1 = pd.read_csv('../Results/resultsimple_buy_sell_at_midpointaccb_nosec_bid_askcontinuous_1action2020-08-09 12:56:18.267022.csv')
midpoint_cont_seed2 = pd.read_csv('../Results/resultsimple_buy_sell_at_midpointaccb_nosec_bid_askcontinuous_1action2020-08-09 12:57:34.024263.csv')

```

### Plot Rewards

```python
fig, axs = plt.subplots(4, 1,figsize=(30, 30))
axs[0].plot(midpoint_disc_seed1['Reward'])
axs[0].title.set_text('midpoint_disc_seed1')

axs[1].plot(midpoint_disc_seed2['Reward'])
axs[1].title.set_text('midpoint_disc_seed2')

axs[2].plot(midpoint_cont_seed1['Reward'])
axs[2].title.set_text('midpoint_cont_seed1')

axs[3].plot(midpoint_cont_seed2['Reward'])
axs[3].title.set_text('midpoint_cont_seed2')
```

```python
fig, axs = plt.subplots(4, 1,figsize=(30, 30))
axs[0].plot(midpoint_disc_seed1['Net_Worth'])
axs[0].title.set_text('midpoint_disc_seed1')

axs[1].plot(midpoint_disc_seed2['Net_Worth'])
axs[1].title.set_text('midpoint_disc_seed2')

axs[2].plot(midpoint_cont_seed1['Net_Worth'])
axs[2].title.set_text('midpoint_cont_seed1')

axs[3].plot(midpoint_cont_seed2['Net_Worth'])
axs[3].title.set_text('midpoint_cont_seed2')
```
